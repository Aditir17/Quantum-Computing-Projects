```python
#This algorithm also states if the function is constant or balanced like Deutche's Algorithm provided number of input qubits can vary(n)

#Necessary imports
import numpy as np
from qiskit import QuantumCircuit,transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
```

```python
#Consatant oracle
n=2
const_oracle=QuantumCircuit(n+1)
output=np.random.randint(2)
if output==1:
    const_oracle.x(n)
const_oracle.draw('mpl')
```

```python
#DJ Circuit
dj_circuit=QuantumCircuit(n+1,n)

#There n number of input qubit initialized to zero and last qubit initialized to 1 
for qubit in range(n):
    dj_circuit.h(qubit)

dj_circuit.x(n)
dj_circuit.h(n)
dj_circuit.barrier()

dj_circuit.compose(const_oracle, inplace=True)
dj_circuit.barrier()

for qubit in range(n):
    dj_circuit.h(qubit)
dj_circuit.barrier()

for i in range(n):
    dj_circuit.measure(i,i)

dj_circuit.draw('mpl')
```

```python
#simulating using qasm simulator
backend=Aer.get_backend('qasm_simulator')

job=transpile(dj_circuit,backend)
circuit=backend.run(job)
result=circuit.result()
counts=result.get_counts(dj_circuit)
print('no of counts:',counts)

plot_histogram(counts)
```

```python
# General Function for the constant Oracle for n-qubits

def dj_oracle_const(n):
    oracle_qc=QuantumCircuit(n+1)
    output= np.random.randint(2)
    if output==1:
        oracle_qc.x(n)
    oracle_gate_const=oracle_qc.to_gate()
    oracle_gate_const.name="Oracle Constant"
    return oracle_gate_const
```

```python
# DJ Algorithm Function

def dj_algorithm(dj_oracle_const,n):
    dj_circuit= QuantumCircuit(n+1,n)
    dj_circuit.x(n)
    dj_circuit.h(n)

    for qubit in range(n):
        dj_circuit.h(qubit)

    dj_circuit.append(dj_oracle_const, range(n+1))

    for qubit in range(n):
        dj_circuit.h(qubit)

    for i in range(n):
        dj_circuit.measure(i,i)

    return dj_circuit
    
```

```python
n=8
oracle_gate_const=dj_oracle_const(n)
dj_circuit= dj_algorithm(oracle_gate_const,n)
dj_circuit.draw('mpl')
```

```python
backend=Aer.get_backend('qasm_simulator')

job=transpile(dj_circuit,backend)
circuit=backend.run(job)
result=circuit.result()
counts=result.get_counts(dj_circuit)
print('no of counts:',counts)

plot_histogram(counts)
```

```python
#general function for balanced oracle

def dj_oracle_balanced(n):
    oracle_qc=QuantumCircuit(n+1)
    b=np.random.randint(1,2**n)
    b_str= format(b,'0'+str(n)+'b')
    
    for qubit in range(len(b_str)):
        if b_str[qubit]=='1':
            oracle_qc.x(qubit)
    for qubit in range(n):
         oracle_qc.cx(qubit,n)
    for qubit in range(len(b_str)):
        if b_str[qubit]=='1':
            oracle_qc.x(qubit)
    oracle_gate_balanced= oracle_qc.to_gate()
    oracle_gate_balance="Oracle Balanced"
    return oracle_gate_balanced
```

```python
# DJ Algorithm Function-Balanced

def dj_algorithm(dj_oracle_balanced,n):
    dj_circuit= QuantumCircuit(n+1,n)
    dj_circuit.x(n)
    dj_circuit.h(n)

    for qubit in range(n):
        dj_circuit.h(qubit)

    dj_circuit.append(dj_oracle_balanced, range(n+1))

    for qubit in range(n):
        dj_circuit.h(qubit)

    for i in range(n):
        dj_circuit.measure(i,i)

    return dj_circuit
```

```python
n=8
oracle_gate_balanced=dj_oracle_balanced(n)
dj_circuit= dj_algorithm(oracle_gate_balanced,n)
dj_circuit.draw('mpl')
```

```python
backend=Aer.get_backend('qasm_simulator')

job=transpile(dj_circuit,backend)
circuit=backend.run(job)
result=circuit.result()
counts=result.get_counts(dj_circuit)
print('no of counts:',counts)

plot_histogram(counts)
```
